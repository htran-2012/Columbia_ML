```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('fire_australia/fire_nrt_V1_96617.csv')
```

```python
df.columns
```

```python
df = df.drop(['satellite', 'instrument', 'version'], axis = 1)
```

```python
df.head()
```

```python
def time_format(num):
    string = '{:04d}'.format(num)
    string = string[:2] + ':' + string[2:]
    return string


df['acq_datetime'] = df.acq_date + ' ' + df.acq_time.apply(time_format)
```

```python
df.head()
```

```python
fig, ax = plt.subplots()
ax.hist(df['latitude'], bins = 50)
plt.plot()
```

```python
fig, ax = plt.subplots()
ax.hist(df['longitude'], bins = 50)
plt.plot()
```

```python
fig, ax = plt.subplots(figsize = (20,20))
ax.scatter(df['latitude'],df['longitude'], alpha = 0.1)
ax.set_xlabel('Latitude')
ax.set_ylabel('Longitude')
plt.plot()
```

```python
df.confidence.value_counts()
```

```python
colors = {'n': 'red', 'h': 'blue', 'l': 'green'}
```

```python
fig, ax = plt.subplots(figsize = (20,20))
ax.scatter(df['latitude'],df['longitude'], alpha = 0.1,  c=df['confidence'].apply(lambda x: colors[x]))
ax.set_xlabel('Latitude')
ax.set_ylabel('Longitude')
plt.plot()
```

```python
df.daynight.value_counts()
```

```python
colors = {'D': 'yellow', 'N': 'orange'}
```

```python
fig, ax = plt.subplots(figsize = (20,20))
ax.scatter(df['latitude'],df['longitude'], alpha = 0.1,  c=df['daynight'].apply(lambda x: colors[x]))
ax.set_xlabel('Latitude')
ax.set_ylabel('Longitude')
plt.plot()
```

```python
fig, ax = plt.subplots(figsize = (18,10))
hb = ax.hexbin(df['latitude'],df['longitude'])
ax.set_xlabel('Latitude')
ax.set_ylabel('Longitude')
ax.set_title('Hexagon binning')
cb = fig.colorbar(hb, ax=ax)
cb.set_label('counts')
plt.plot()
```

```python
day = df['daynight'] == 'D'
night = df['daynight'] == 'N'
```

```python

fig, ax = plt.subplots(figsize = (20,20))
ax.scatter(df.loc[day,'latitude'],df.loc[day, 'longitude'], alpha = 0.1,  label = 'Day', marker= 'o')
ax.scatter(df.loc[night,'latitude'],df.loc[night, 'longitude'], alpha = 0.1,  label = 'night', marker = '^')
ax.set_xlabel('Latitude')
ax.set_ylabel('Longitude')
ax.legend()
plt.plot()
```
